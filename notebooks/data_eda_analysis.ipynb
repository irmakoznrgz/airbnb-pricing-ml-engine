```python
cols_to_keep = [
    'neighbourhood_cleansed',
    'latitude', 
    'longitude',
    'property_type', 
    'room_type',
    'accommodates',
    'bathrooms_text', 
    'bedrooms', 
    'beds',
    'amenities',
    'minimum_nights', 
    'maximum_nights',
    'availability_365',
    'number_of_reviews', 
    'review_scores_rating',
    'host_is_superhost',
    'host_identity_verified',
    'calculated_host_listings_count',
    'instant_bookable',
    'reviews_per_month',
    'review_scores_location',
    'review_scores_cleanliness',
    'price'
]
```

```python
df= df[cols_to_keep]
```

```python
df = df.dropna(subset=['price'])
df = df.drop(columns=['instant_bookable'])
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)
df['review_scores_rating'] = df['review_scores_rating'].fillna(df['review_scores_rating'].median())
df['bedrooms'] = df['bedrooms'].fillna(df.groupby('room_type')['bedrooms'].transform('median'))
df['beds'] = df['beds'].fillna(df.groupby('accommodates')['beds'].transform('median'))
df['bathrooms_text'] = df['bathrooms_text'].fillna(df['bathrooms_text'].mode()[0])
df['minimum_nights'] = df['minimum_nights'].fillna(df['minimum_nights'].median())
df['maximum_nights'] = df['maximum_nights'].fillna(df['maximum_nights'].median())
df['review_scores_location'] = df['review_scores_location'].fillna(df['review_scores_location'].median())
df['review_scores_cleanliness'] = df['review_scores_cleanliness'].fillna(df['review_scores_cleanliness'].median())
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(x=df['price'])
plt.title('Price Breakdown Before Cleaning')

plt.subplot(1, 2, 2)
sns.boxplot(x=df['minimum_nights'])
plt.title('Minimum Nights Breakdown Before Cleaning')
plt.show()
```

```python
def remove_outliers_iqr(dataframe, column):
    Q1 = dataframe[column].quantile(0.25)
    Q3 = dataframe[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    if lower_bound < 0:
        lower_bound = 0
        
    df_filtered = dataframe[(dataframe[column] >= lower_bound) & (dataframe[column] <= upper_bound)]
    
    print(f"Lower Bound for {column}: {lower_bound} | Upper Bound: {upper_bound}")
    return df_filtered

print("Data shape (Before Cleaning):", df.shape)
df = remove_outliers_iqr(df, 'price')

df = remove_outliers_iqr(df, 'minimum_nights')
print("Data shape (After Cleaning):", df.shape)
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(x=df['price'])
plt.title('Price Breakdown After Cleaning')

plt.subplot(1, 2, 2)
sns.boxplot(x=df['minimum_nights'])
plt.title('Minimum Nights Breakdown After Cleaning')
plt.show()
```

```python
df['bathrooms'] = df['bathrooms_text'].str.extract('([0-9.]+)', expand=False).astype(float)
df['bathrooms'] = df['bathrooms'].fillna(0.5)
df = df.drop(columns=['bathrooms_text'])

tf_columns = ['host_is_superhost', 'host_identity_verified']
for col in tf_columns:
    df[col] = df[col].map({'t': 1, 'f': 0})
    
df['has_wifi'] = df['amenities'].str.contains('Wifi|Internet', case=False, na=False).astype(int)
df['has_ac'] = df['amenities'].str.contains('Air conditioning', case=False, na=False).astype(int)
df['has_pool'] = df['amenities'].str.contains('Pool', case=False, na=False).astype(int)
df['has_view'] = df['amenities'].str.contains('view|sea|bosphorus', case=False, na=False).astype(int)
df['has_parking'] = df['amenities'].str.contains('parking|garage', case=False, na=False).astype(int)
df['has_elevator'] = df['amenities'].str.contains('elevator', case=False, na=False).astype(int)
df['is_pet_friendly'] = df['amenities'].str.contains('pets allowed', case=False, na=False).astype(int)
df['has_balcony'] = df['amenities'].str.contains('balcony|terrace|patio', case=False, na=False).astype(int)

df = df.drop(columns=['amenities'])

df.shape
df.head(3)
```

```python
type_counts = df['property_type'].value_counts()

rare_types = type_counts[type_counts < 100].index

df['property_type'] = df['property_type'].replace(rare_types, 'Other')

print(df['property_type'].value_counts())
```

```python
type_counts = df['neighbourhood_cleansed'].value_counts()

rare_types = type_counts[type_counts < 200].index

df['neighbourhood_cleansed'] = df['neighbourhood_cleansed'].replace(rare_types, 'Other')

print(df['neighbourhood_cleansed'].value_counts())
```

```python
import os

categorical_cols = ['neighbourhood_cleansed', 'property_type', 'room_type']
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True, dtype=int)

output_dir = "../data/processed"
os.makedirs(output_dir, exist_ok=True)

output_path = f"{output_dir}/cleaned_listings.csv"
df.to_csv(output_path, index=False)
```
